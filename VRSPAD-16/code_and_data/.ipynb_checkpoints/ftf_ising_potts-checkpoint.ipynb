```python
import pynq
import time
import numpy
np = numpy
from matplotlib import pyplot as plt
import math
import bm_utils

qrng_bm_overlay = pynq.Overlay("design_1.bit")

IP_BASE_ADDRESS = qrng_bm_overlay.ip_dict["big_gpio_0/S_AXI"]['phys_addr']
ADDRESS_RANGE = qrng_bm_overlay.ip_dict["big_gpio_0/S_AXI"]['addr_range']

mmio = pynq.MMIO(IP_BASE_ADDRESS, ADDRESS_RANGE)

#important: define what SPAD bias voltage is being used: external power supply setting should be reflected here
vspad = 29
```

```python
def load_weights(w, b):
    #loads values into scan FPGA scan chain
    read_weights = numpy.zeros([16, 16])
    read_biases = numpy.zeros([16])
    for i in range(15,-1,-1):
        for j in range(15,-1,-1):
            mmio.write(1*4, 0)
            mmio.write(0, int(w[i,j]))
            read_weights[i,j] = mmio.read(4*(10+99))
            mmio.write(1*4, 1)
        mmio.write(1*4, 0)
        mmio.write(0, int(b[i]))
        read_biases[i] = mmio.read(4*(10+99))
        mmio.write(1*4, 1)
    return read_weights, read_biases

def read_rates(print_measurements=False):
    #reads SPAD triggering rates and BM state change rates
    mmio.write(4*5, 1) #manual reset
    mmio.write(4*5, 0) 
    start_t = time.time()
    mmio.write(4*7,1)
    time.sleep(0.1)
    total_t = time.time()-start_t
    mmio.write(4*7,0)
    
    raw_rate = np.zeros([16])
    clk_rate = np.zeros([16])
    flip_rate = np.zeros([16])
    for i in range(16):
        raw_rate[i] = 1e-6/total_t*mmio.read((i+48+10)*4)
        clk_rate[i] = 1e-6/total_t*mmio.read((i+0+10)*4)
        flip_rate[i] = 1e-6/total_t*mmio.read((i+16+10)*4)
        if print_measurements:
            print("channel %i: %.2fMHz raw, %.2fMHz synchronized, %.2fMHz flip rate"%(i,raw_rate[i], clk_rate[i], flip_rate[i]))
    return clk_rate, flip_rate, raw_rate

def read_sums(): #for checking that hardware/verilog is working correctly
    for i in range(16):
        print(mmio.read((i+32+10)*4))
        
def set_light(intensity):
    mmio.write(4*11, intensity)
        
def set_light_decriment_period(period): #during a DMA transfer, the illumination decriments after every 2^period DMA transfers 
    mmio.write(4*12, period) 
        
def set_update_delay(delay):
    mmio.write(4*9, delay) #time to allow SPADs to update
        
def set_ising_mode(): 
    mmio.write(4*2, 0)
def set_potts_mode():
    mmio.write(4*2, 1)
        
def basic_bias_cal(cal_file, zero_firing_rate):
    #loads a calibration file, and finds correct bias offsets for each SPAD so that the firing rate is identical
    cal_data = numpy.loadtxt(cal_file, delimiter=",")
    cal_data[:,0:75] = 0 #set some stuff to zero so that it does not falsely get registered as the proper calibration point
    zero_points = numpy.argmin((cal_data-zero_firing_rate)**2, axis=1)
    
    #finds slope/ratio/temperature of each SPAD around the zero points:
    slope_ratios = np.zeros([16])
    for i in range(16):
        slope_ratios[i] = cal_data[i,100]/cal_data[i, 200]
        
    SPAD_temps = 101./np.log(slope_ratios) #100, becuase we measured the slope over a 100 DAC-step long segment
    temp = np.median(SPAD_temps)
    print(SPAD_temps)
    return zero_points, temp

#global resources for performing DMA transfers
xlnk = pynq.Xlnk()
bm_states = xlnk.cma_array(shape=(int(1e5),), dtype=np.uint16)
mmio.write(10*4, int(1e5)) #set the DMA transfer size in hardware
dma_fromPL = qrng_bm_overlay.axi_dma_0

def get_raw_states(downsample=0):
    #reads consecutive states from the Boltzmann machine
    mmio.write(6*4, downsample) #set downsampling before starting xfer: number of clock cycles to wait between samples
    dma_fromPL.recvchannel.transfer(bm_states)
    dma_fromPL.recvchannel.wait()
    raw_states = (((bm_states[:,None] & (1 << np.arange(16)))) > 0).astype(int)
    return raw_states
```

```python
rate_cal_data = np.zeros([16,256,9])
set_update_delay(50000)#very long, so that we can get clean reading of rates
set_light_decriment_period(20) #long enough that the light is constant over the course of sampling
plt.figure(figsize=(12,12))
for illum in range(9):
    #set illumination on SPAD:
    set_light(illum)
    for b in range(0,256):
        print('%i    '%b, end='\r')#dumb output of progress
        load_weights(np.zeros([16,16])+0, np.zeros([16])+b)
        time.sleep(0.1)
        rate_cal_data[:,b,illum], _, _ = read_rates()
        
    plt.subplot(3, 3, illum+1)
    for chan in range(16):
        plt.plot(range(256),rate_cal_data[chan,:,illum])
        
    plt.title('SPAD triggering rates, illum=%i'%illum)
    plt.xlabel('DAC value')
    plt.ylabel('triggering rate, MHz')
    plt.yscale('log')
plt.tight_layout()
plt.show()

set_light(0)
    
for illum in range(9):
    numpy.savetxt('SPAD_rates_%iV_illum%i.csv'%(vspad, illum), rate_cal_data[:,:,illum], delimiter=',')
```

```python
#test neurons individually
set_points = [[2, 0.05],[14, 0.3]] #for 29V. illumination, middlepoint rate
set_points = [[1, 0.2],[8, 4.]] #for 31V
plt.figure(figsize=(10,10))

set_update_delay(25)
set_ising_mode()
set_light_decriment_period(20) #long enough that the light is constant over the course of sampling

test_range_width = 75
test_range = range(-test_range_width,test_range_width+1)
ising_tanh_curves = numpy.zeros([test_range_width*2+1, 8, len(set_points)])

temps = []

for k, set_point in enumerate(set_points):
    bias_zero_points, temp = basic_bias_cal("SPAD_rates_%iV_illum%i.csv"%(vspad, set_point[0]), set_point[1])
    set_light(set_point[0])
    temps.append(temp)
    
    for i in test_range:
        print("%i    "%i, end='\r')#dumb output to indicate that code is actually progressing
        biases = numpy.array([1, -1]*8)*i + bias_zero_points
            #second entry in [1, -1] is the 'positive' ising SPAD selector, and should have more negative DAC value
            #to acheive higher firing rate and thus higher probability when the input i is positive
        load_weights(np.zeros([16,16])+0, biases)
        raw_states = get_raw_states(downsample=250)
        ising_states = bm_utils.raw2ising(raw_states)
        ising_tanh_curves[i+test_range_width, :, k] = np.mean(ising_states, axis=0)

    plt.plot(test_range, 0.5*(numpy.tanh(np.array(test_range)/temp)+1.), linewidth=5.0)
    for i in range(8):
        plt.plot(test_range, ising_tanh_curves[:,i,k])
        
    numpy.savetxt("ising_tanh_curves_%iV_illum%i.csv"%(vspad, set_point[0]), ising_tanh_curves[:,:,k], delimiter=',')

plt.xlabel("Energy summation")
plt.ylabel("Probability of '1' state")
plt.title("Probability transfer functions of ising nodes, temps = %.1f,%.1f"%(temps[0], temps[1]))
plt.show()
```

```python
#great! Now that we have a bunch of working nodes, time to create a Boltzmann machine and test it:
import importlib
importlib.reload(bm_utils) #used during development to load in changes (fixes) made in other file

set_points = [[2, 0.05],[14, 0.3]] #for 29V. illumination, middlepoint rate
# set_points = [[1, 0.2],[8, 4.]] #for 31V

weights, biases, energies = bm_utils.eight_bit_ising()
set_light_decriment_period(20) #long enough that the light is constant over the course of sampling

def setup_ising():
    set_ising_mode()
    hw_weights, hw_biases = bm_utils.ising2hardware(weights, biases)
    load_weights(hw_weights, hw_biases+bias_zero_points)
    
def ising_decode(raw_states):
    ising_states = bm_utils.raw2ising(raw_states)
    return bm_utils.ising2index(ising_states)

def setup_potts():
    #run potts machine on hardware:
    set_potts_mode()
    potts_weights, potts_biases = bm_utils.ising2potts(weights, biases)
    hw_weights, hw_biases = bm_utils.potts2hardware(potts_weights, potts_biases)
    load_weights(hw_weights, hw_biases+bias_zero_points)
    
def potts_decode(raw_states):
    potts_states = bm_utils.raw2potts(raw_states)
    return bm_utils.potts2index(potts_states)

def measure(theory_probs, n2, decoder):
    hist = np.zeros([2**8])
    kl_div = np.zeros([n2])
    for i in range(n2):
        raw_states = get_raw_states(downsample=10)
#         print(raw_states[0:5,:])
        bm_states = decoder(raw_states[10:,:]) #for some reason (FIFO in DMA?), first few slots are not populated correctly

        hist = hist + np.bincount(bm_states, minlength=2**8)
        measured_probs = hist/((i+1)*bm_states.shape[0])
        kl_div[i] = np.sum(theory_probs*np.log2(theory_probs/(measured_probs+1e-16)))
        if i%1 == 0:
            print("%i: %.6f"%(i,kl_div[i]), end='\r')
            
    return kl_div, measured_probs

kl_div_axis = numpy.linspace(0.1, 5, 50) #in millions

#run ising machine on hardware:
for k, set_point in enumerate(set_points):
    bias_zero_points, temp = basic_bias_cal("SPAD_rates_%iV_illum%i.csv"%(vspad, set_point[0]), set_point[1])
    set_light(set_point[0])
    set_update_delay(25) #after a state change, BM holds for 25 clock cycles while DACs are updated

    theory_probs = np.exp(-energies/temp)
    theory_probs = theory_probs/np.sum(theory_probs)

    setup_ising()
    kl_div_ising, measured_probs_ising = measure(theory_probs, 50, ising_decode)
    setup_potts()
    kl_div_potts, measured_probs_potts = measure(theory_probs, 50, potts_decode)

    numpy.savetxt("probabilities_%iV_illum%i.csv"%(vspad, set_point[0]), [theory_probs, measured_probs_ising, measured_probs_potts], delimiter=',')
    numpy.savetxt("kl_div_%iV_illum%i.csv"%(vspad, set_point[0]), [kl_div_ising, kl_div_potts], delimiter=',')

    print("plotting")
    plt.figure()
    plt.plot(kl_div_axis, kl_div_ising)
    plt.plot(kl_div_axis, kl_div_potts)
    plt.xlabel("Millions of Samples")
    plt.ylabel("KL divergence between HW BM and theory")
    plt.legend(['Ising', 'Potts'])
    plt.yscale('log')
    plt.xscale('log')
    plt.show()

    width=0.35
    plt.figure(figsize=(8, 3))
    plot_indices = np.argsort(theory_probs)
    plt.plot(measured_probs_ising[plot_indices])
    plt.plot(measured_probs_potts[plot_indices])
    plt.plot(theory_probs[plot_indices])
    plt.title("State probabilties, ordered by ideal probability")
    plt.xlabel("BM state")
    plt.ylabel("State Probability")
    plt.tight_layout()
    plt.show()
```

```python
#Annealing!  By making the illumination decriment periodically while sampling, a linear annealing schedule is acheived.

#make a new BM: lets us easily see how results change with different weights
weights, biases, energies = bm_utils.eight_bit_ising()

def setup_ising():
    set_ising_mode()
    hw_weights, hw_biases = bm_utils.ising2hardware(weights, biases)
    load_weights(hw_weights, hw_biases+bias_zero_points)

def setup_potts():
    #run potts machine on hardware:
    set_potts_mode()
    potts_weights, potts_biases = bm_utils.ising2potts(weights, biases)
    hw_weights, hw_biases = bm_utils.potts2hardware(potts_weights, potts_biases)
    load_weights(hw_weights, hw_biases+bias_zero_points)
    
set_light(14)
per = 7
set_light_decriment_period(per) #light decriments every 2^period DMA transfer cycles.  Full DMA transfer is ~2^16.6
set_update_delay(25)

bias_zero_points, temp = basic_bias_cal("SPAD_rates_%iV_illum%i.csv"%(vspad, 0), 0.01) #calibrate based on end illumination 

setup_ising()
states_ising = get_raw_states(downsample=10)
states_ising = ising_decode(states_ising[10:15*2**per])
setup_potts()
states_potts = get_raw_states(downsample=10)
states_potts = potts_decode(states_potts[10:15*2**per])

set_light(0)

numpy.savetxt("annealing_%iV.csv"%vspad, [energies[states_ising], energies[states_potts]], delimiter=',')

plt.figure(figsize=(10,10))
plt.subplot(2, 1, 1)
plt.plot(energies[states_ising])
plt.xlabel('sampling timestep')
plt.ylabel('Energy of model state')
plt.title('Ising model')

plt.subplot(2, 1, 2)
plt.plot(energies[states_potts])
plt.xlabel('sampling timestep')
plt.ylabel('Energy of model state')
plt.title('Potts model')

plt.show()
```
